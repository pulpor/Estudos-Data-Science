pandas é uma biblioteca,
funciona com dataframes (uma tabela dentro do pandas)

```python
#importando pandas 

import pandas as pd
```

```python
dataframe = pd.DataFrame()   #dataframe vazio
```

```python
#criando um dataframe a partir de um dicionario

venda = {   'data': ['03/02/2023', '01/01/2023'],
            'valor': [500,300],
            'produto': ['feijao', 'arroz'],
            'quantidade': [50,70]
        }

print(venda)
```

```python
vendas_df = pd.DataFrame(venda)

print(vendas_df)

display(vendas_df)
```

```python
#importando arquivos externos, excel

vendas_df = pd.read_excel("assets/Vendas.xlsx")

display(vendas_df)
```

```python
#metodo de visualização de dados: 1) head 2) shape e 3) describe

display(vendas_df.head(15))  #head ve o inicio da tabela
```

```python
print(vendas_df.shape) #shape nao tem abre e fecha parenteses, ele mostra quantas linhas e quantas colunas tem 
```

```python
display(vendas_df.describe()) #describe te da o resumo da tabela
```

```python
#pegar 1 coluna ou mais de uma [um colchete por coluna]

produtos = vendas_df[['Produto','ID Loja']]
display(produtos)
```

```python
#pegar linhas especificas

display(vendas_df.loc[1:10])
```

```python
#pegar linhas que correspondem a uma condição

parametro = vendas_df['ID Loja'] == 'Norte Shopping'

display(vendas_df.loc[parametro])
```

```python
#pegar linhas que correpondem a uma condição + colunas que eu acho relevante

                                    #padrão do loc é = 'linhas' , ['coluna1', 'coluna2', 'coluna3']
parametro = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]

display(parametro)
```
