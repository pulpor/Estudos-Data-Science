```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import quandl
import mplcyberpunk
```

```python
acao = 'QUAL3.SA'


acao_data = yf.download(acao, start='2021-01-01')
cdi_data = quandl.get('BCB/11', start_date='2021-01-01')

#normalize
returns = acao_data['Adj Close'].pct_change().dropna()
cdi_returns = cdi_data['Value']/100

#calcula o retorno acumulado
acao_cumulative = (1 + returns).cumprod() - 1
cdi_cumulative = (1 + cdi_returns).cumprod() - 1

#plotar grafico
plt.style.use("cyberpunk")
plt.plot(acao_cumulative, label='Ação')
plt.plot(cdi_cumulative, label='CDI')
plt.legend()
plt.title('Retorno Acumulado da Ação e do CDI')
plt.show()


```
